```python
retail.columns = retail.columns.str.lower()
```

```python
retail.head()
```

```python
retail.columns = retail.columns.str.replace(' ', '_')
```

```python
retail.head(5)
```

```python
retail['invoicedate'] = pd.to_datetime(retail['invoicedate'])
```

```python
retail['country'] = retail['country'].str.upper()
```

```python
retail.head()
```

```python
retail['customerid'] = retail['customerid'].astype(int)
```

```python
retail['revenue'] = retail['quantity'] * retail['unitprice']

retail['month'] = retail['invoicedate'].dt.to_period('M')
```

```python
retail.head()
```

```python
import matplotlib.pyplot as plt
monthly_sales = retail.groupby('month')['revenue'].sum()

monthly_sales.plot()

plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.hist(retail['quantity'])

plt.title('Quantity Distribution')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(retail['quantity'])

plt.title('Quantity Box Plot')
plt.ylabel('Quantity')
plt.show()
```

```python
top_products = retail.groupby('description')['quantity'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar')

plt.title('Top 10 Selling Products')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
top_countries = retail.groupby('country')['revenue'].sum().sort_values(ascending=False).head(5)

top_countries.plot(kind='pie')

plt.title('Revenue Share by Country')
plt.ylabel('')
plt.show()
```
